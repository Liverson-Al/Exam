```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import axes3d
import matplotlib as mpl

ax = axes3d.Axes3D(plt.figure())
i = np.arange(-10,10,0.01)
X, Y = np.meshgrid(i,i)
Z = X**2-Y**2
ax.plot_wireframe(X, Y, Z, rstride = 25, cstride = 100)


plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
```

```python
x = y = np.linspace(-3,3,74)
X, Y = np.meshgrid(x,y)
R = np.sqrt(X**2+Y**2)
Z = np.sin(4*R)/R
fig, ax = plt.subplots(1,3, figsize=(14,4), subplot_kw = dict(projection = '3d'))

#subplot0 - surfce

norm = mpl.colors.Normalize(-abs(Z).max(), abs(Z).max())

p = ax[0].plot_surface(X,Y,Z, linewidth = 0, antialiased = False, norm = norm, cmap = mpl.cm.Blues)
cb = fig.colorbar(p, ax = ax[0], shrink = 0.6)

ax[0].set_xlabel("$x$", fontsize = 16)
ax[0].set_ylabel("$y$", fontsize = 16)
ax[0].set_zlabel("$z$", fontsize = 16)

#subplot1 - wireframe

ax[1].plot_wireframe(X,Y,Z, rstride = 3, cstride=3, color="darkgrey")
ax[1].set_title("plot_wireframe")

ax[1].set_xticks([])
ax[1].set_yticks([])
ax[1].set_zticks([])

#subplot2 -    
ax[2].contour(X,Y,Z, zdir = "z", offset=-1, norm = norm, cmap=mpl.cm.jet)
ax[2].contour(X,Y,Z, zdir = "y", offset=3, norm = norm, cmap=mpl.cm.Blues)

bbox_min = np.min([X,Y,Z])
bbox_max = np.max([X,Y,Z])
ax[0].auto_scale_xyz([bbox_min, bbox_max],[bbox_min, bbox_max],[bbox_min, bbox_max])
                  
plt.show()
fig.savefig("5b0.png", dpi = 100)
```

```python
def makeData():
  x = np.arange(-10,10,0.1)
  y = np.arange(-10,10,0.1)
  xgrid, ygrid = np.meshgrid(x,y)
  zgrid = np.sin(xgrid)*np.sin(ygrid)/(xgrid*ygrid)
  return xgrid, ygrid, zgrid

x,y,z = makeData()

fig=plt.figure()
ax = axes3d(fig)

ax.meshgrid(x,y,z)

```

```python
import seaborn as sns
import pandas as pd

np.random.seed(1234)

v1=pd.Series(np.random.normal(0,10,1000),name = 'v1')
v2=pd.Series(2*v1 + np.random.normal(60,15,1000),name = 'v2')

plt.figure()
plt.hist(v1, alpha=0.7,bins = np.arange(-50,150,5), label='v1')
plt.hist(v2, alpha=0.7,bins = np.arange(-50,150,5), label='v2')

plt.legend()
plt.show()
```
