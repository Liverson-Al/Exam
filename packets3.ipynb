```python
!head stockholm_tmp.dat

data = np.genfromtxt('stockholm_tmp.dat')

import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize = (14,4))
#ax.plot(data[:,0]+data[:,1]/12.0+data[:,2]/365, data[:,3])
dr = 0.0
counter = 0
middle = []
counter_month = 0
for i in range(1,np.size(data[:,3])-1):
  if (data[i,1]==data[i-1,1]):
    dr += data[i-1, 3]
    counter+=1
  else:
    middle.append(dr/counter)
    dr = 0.0
    counter = 0
    counter_month+=1
print(middle)

ax.plot(list(range(0, len(middle))),middle[:])

ax.axis('tight')
ax.set_xlabel('year')
ax.set_ylabel('temperature')


```
